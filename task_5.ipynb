```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure plots display inline
%matplotlib inline

# Load Titanic dataset
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
print("Dataset loaded successfully. Shape:", df.shape)
print(df.head())
```

```python
age_data = df['Age'].dropna()
print("Using distplot to visualize age distribution...")
sns.distplot(age_data, bins=30, kde=True, color='teal')
plt.title('Age Distribution (distplot)')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

```python
print("Histogram of Age distribution...")
sns.histplot(df['Age'].dropna(), bins=30, kde=False, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
print("KDE plot of Age...")
sns.kdeplot(df['Age'].dropna(), shade=True, color='purple')
plt.title('Age Density')
plt.xlabel('Age')
plt.show()
```

```python
print("Countplot of Gender...")
sns.countplot(x='Sex', data=df, palette='Set2')
plt.title('Gender Count')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
print("Barplot of Survival Rate by Class...")
sns.barplot(x='Pclass', y='Survived', data=df, palette='Blues')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
print("Boxplot of Age by Class...")
sns.boxplot(x='Pclass', y='Age', data=df, palette='Pastel1')
plt.title('Age Distribution by Class')
plt.xlabel('Class')
plt.ylabel('Age')
plt.show()
```

```python
print("Violinplot of Age by Gender...")
sns.violinplot(x='Sex', y='Age', data=df, palette='Set3')
plt.title('Age Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

```python
print("Scatterplot of Age vs Fare...")
sns.scatterplot(x='Age', y='Fare', data=df, hue='Survived', palette='coolwarm')
plt.title('Age vs Fare (Survival Highlighted)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
print("Lineplot of Survival Rate by Age...")
age_survival = df.groupby('Age')['Survived'].mean().dropna().reset_index()
sns.lineplot(x='Age', y='Survived', data=age_survival, color='green')
plt.title('Survival Rate by Age')
plt.xlabel('Age')
plt.ylabel('Survival Rate')
plt.show()
```

```python
print("Heatmap of numerical correlations...")
corr_matrix = df[['Survived', 'Age', 'Fare', 'Pclass']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu')
plt.title('Correlation Heatmap')
plt.show()
```

```python
print("Pairplot of selected features...")
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass']].dropna(), hue='Survived', palette='husl')
plt.suptitle('Pairplot of Titanic Features', y=1.02)
plt.show()
```

```python
print("Jointplot of Age vs Fare...")
sns.jointplot(x='Age', y='Fare', data=df, kind='scatter', hue='Survived', palette='coolwarm')
plt.show()
```
