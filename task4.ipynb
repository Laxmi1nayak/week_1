```python
import numpy as np
import matplotlib.pyplot as plt

# Ensure plots display inline (for Jupyter/Colab)
%matplotlib inline
print("Libraries imported successfully.")
```

```python
# Months of the year
import numpy as np
months = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                   'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# Monthly sales for three products (in thousands)
product_A = np.array([12, 15, 14, 17, 20, 22, 21, 19, 18, 23, 25, 28])
product_B = np.array([8, 9, 10, 12, 13, 15, 14, 13, 12, 14, 16, 18])
product_C = np.array([5, 6, 7, 6, 8, 9, 10, 11, 10, 12, 13, 15])
print("Months:", months)
print("Product A Sales:", product_A)
print("Product B Sales:", product_B)
print("Product C Sales:", product_C)
```

```python
# ðŸ“Œ Step 3: Line Plot â€“ Monthly Sales Trend
print("Generating line plot for monthly sales trends...")

plt.figure(figsize=(10, 6))
plt.plot(months, product_A, label='Product A', marker='o', color='blue')
plt.plot(months, product_B, label='Product B', marker='s', color='green')
plt.plot(months, product_C, label='Product C', marker='^', color='red')

plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales (in thousands)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ðŸ“Œ Step 4: Bar Plot â€“ December Sales Comparison
print("Generating bar plot for December sales comparison...")

december_sales = np.array([product_A[-1], product_B[-1], product_C[-1]])
products = ['Product A', 'Product B', 'Product C']

print("December Sales:", december_sales)

plt.figure(figsize=(8, 5))
plt.bar(products, december_sales, color=['blue', 'green', 'red'])

plt.title('December Sales Comparison')
plt.ylabel('Sales (in thousands)')
plt.show()
```

```python
# ðŸ“Œ Step 5: Scatter Plot â€“ Product A vs Product B
print("Generating scatter plot for Product A vs Product B...")

plt.figure(figsize=(8, 5))
plt.scatter(product_A, product_B, color='purple', marker='x')

plt.title('Product A vs Product B Sales')
plt.xlabel('Product A Sales')
plt.ylabel('Product B Sales')
plt.grid(True)
plt.show()
```

```python
# ðŸ“Œ Step 6: Image Plot â€“ Sales Heatmap
print("Generating heatmap of sales data...")

sales_matrix = np.array([product_A, product_B, product_C])
print("Sales Matrix:\n", sales_matrix)

plt.figure(figsize=(10, 4))
plt.imshow(sales_matrix, cmap='viridis', aspect='auto')
plt.colorbar(label='Sales (in thousands)')
plt.title('Sales Heatmap')
plt.xlabel('Month Index')
plt.ylabel('Product Index (0=A, 1=B, 2=C)')
plt.show()
```

```python
# ðŸ“Œ Step 7: Subplots â€“ All in One View
print("Generating subplot with all visualizations...")

fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Line plot
axs[0, 0].plot(months, product_A, label='Product A', color='blue')
axs[0, 0].set_title('Product A Sales')
axs[0, 0].set_ylabel('Sales')
axs[0, 0].grid(True)

# Bar plot
axs[0, 1].bar(products, december_sales, color=['blue', 'green', 'red'])
axs[0, 1].set_title('December Sales')

# Scatter plot
axs[1, 0].scatter(product_A, product_B, color='purple')
axs[1, 0].set_title('A vs B Sales')
axs[1, 0].set_xlabel('Product A')
axs[1, 0].set_ylabel('Product B')

# Heatmap
im = axs[1, 1].imshow(sales_matrix, cmap='coolwarm', aspect='auto')
fig.colorbar(im, ax=axs[1, 1])
axs[1, 1].set_title('Sales Heatmap')

plt.tight_layout()
plt.show()
```
