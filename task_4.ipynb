```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Ensure plots display inline
%matplotlib inline

# Load Titanic dataset
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
print("Dataset loaded successfully. Shape:", df.shape)
print(df.head())
```

```python
# Group by age and calculate survival rate
age_survival = df.groupby('Age')['Survived'].mean().dropna()

print("Survival rate by age:\n", age_survival.head())

plt.figure(figsize=(10, 5))
plt.plot(age_survival.index, age_survival.values, color='blue', marker='o')
plt.title('Survival Rate by Age')
plt.xlabel('Age')
plt.ylabel('Survival Rate')
plt.grid(True)
plt.show()
```

```python
pclass_counts = df['Pclass'].value_counts().sort_index()
print("Passenger class counts:\n", pclass_counts)

plt.figure(figsize=(6, 4))
plt.bar(pclass_counts.index.astype(str), pclass_counts.values, color='green')
plt.title('Passenger Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
print("Scatter plot of Age vs Fare...")

plt.figure(figsize=(8, 5))
plt.scatter(df['Age'], df['Fare'], alpha=0.5, color='purple', marker='x')
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```

```python
print("Histogram of Age distribution...")

plt.figure(figsize=(8, 5))
plt.hist(df['Age'].dropna(), bins=30, color='orange', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
gender_counts = df['Sex'].value_counts()
print("Gender counts:\n", gender_counts)

plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['lightblue', 'lightpink'])
plt.title('Gender Proportion')
plt.show()
```

```python
# Create pivot table
heatmap_data = df.pivot_table(index='Sex', columns='Pclass', values='Survived', aggfunc='mean')
print("Survival heatmap data:\n", heatmap_data)

plt.figure(figsize=(6, 4))
plt.imshow(heatmap_data, cmap='coolwarm', aspect='auto')
plt.colorbar(label='Survival Rate')
plt.title('Survival Rate by Gender and Class')
plt.xlabel('Pclass')
plt.ylabel('Sex')
plt.xticks(ticks=[0, 1, 2], labels=['1st', '2nd', '3rd'])
plt.yticks(ticks=[0, 1], labels=heatmap_data.index)
plt.show()
```

```python
print("Generating subplots...")

fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Line plot
axs[0, 0].plot(age_survival.index, age_survival.values, color='blue')
axs[0, 0].set_title('Survival Rate by Age')
axs[0, 0].set_xlabel('Age')
axs[0, 0].set_ylabel('Rate')

# Bar plot
axs[0, 1].bar(pclass_counts.index.astype(str), pclass_counts.values, color='green')
axs[0, 1].set_title('Passenger Class Distribution')
axs[0, 1].set_xlabel('Class')
axs[0, 1].set_ylabel('Count')

# Scatter plot
axs[1, 0].scatter(df['Age'], df['Fare'], alpha=0.5, color='purple')
axs[1, 0].set_title('Age vs Fare')
axs[1, 0].set_xlabel('Age')
axs[1, 0].set_ylabel('Fare')

# Histogram
axs[1, 1].hist(df['Age'].dropna(), bins=30, color='orange', edgecolor='black')
axs[1, 1].set_title('Age Distribution')
axs[1, 1].set_xlabel('Age')
axs[1, 1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
